```python
import pandas as pd
import seaborn 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('/workspaces/Data-analysis1/salary_years.csv')
# x=data['years']
# print(x)
```

```python
seaborn.lmplot(x='years',y='salary', data=data)
plt.show()
```

```python
fit=sm.formula.ols('salary~years',data=data).fit()
print(fit.params)
```

```python
seaborn.lmplot(x='years',y='salary', data=data, fit_reg=True)
fit = sm.formula.ols('salary ~ years', data=data).fit()
print(fit.params)
plt.show()
```

```python
from scipy . optimize import leastsq
def error(params, x, y):
    a, b = params
    return y - (a * x + b)

x = data['years']
y = data['salary']
# print(x)
p0 = [0, 1]
Para = leastsq(error, p0, args=(x,y))
a, b = Para[0]
print(a,b)
```
